```python
from pyECLAT import ECLAT as el
```

```python
ecl=el(data=df)
ecl.df_bin
```

```python
items_total=ecl.df_bin.astype(int).sum(axis=0)
items_total
```

```python
df=pd.DataFrame({'items':items_total.index,
                 'transactions':items_total.values})
df
```

```python
# Number of items in each transactions
items_per_transaction=ecl.df_bin.astype(int).sum(axis=1)
items_per_transaction
```

## Implementation ECLAT

```python
min_support=0.001
min_combination=2
max_combination=max(items_per_transaction)
```

```python
rules_indices, rule_support=ecl.fit(min_support=min_support,
                                   separator='&',
                                   verbose=True,
                                   min_combination=min_combination,

                                   )
```

```python
result=pd.DataFrame(rule_support.items(),
                     columns=['items','support']
                     )
result.sort_values(by=['support'],ascending=False)
result
```
